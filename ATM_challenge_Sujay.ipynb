```python
#ATM challenge

#Check balance function
def check_balance():
  print(f"Your current balance is: £{balance:.2f}")
  while True:
    exit = input("Would you like to perform another  action? Y/N")
    if exit == "N":
      print("Exiting ATM")
      break
    elif exit == "Y":
      return menu()
    else:
      print("Invalid option, try again")

#Deposit function
def deposit():
  global balance
  deposit_amount = int(input("Enter the amount you would like to deposit into your account:"))
  balance = balance + deposit_amount
  print(f"The amount of: £{deposit_amount:.2f} has been successfully deposited into your account")
  while True:
    exit = input("Would you like to perform another  action? Y/N")
    if exit == "N":
      print("Exiting ATM")
      break
    elif exit == "Y":
      return menu()
    else:
      print("Invalid option, try again")

#Withdraw function
def withdraw():
  global balance
  choice = int(input("Please select the amount you would like to withdraw:\n"\
                     "1: £10\n"
                     "2: £20\n"
                     "3: £50\n"
                     "4: £100\n"
                     "5: £500\n"
                     "6: Other amount"))
  if choice == 1:
    if balance >= 10:
      balance = balance - 10
      print("Transaction successful")
    else:
      print("Insufficient Balance")
  elif choice == 2:
    if balance >= 20:
      balance = balance - 20
      print("Transaction successful")
    else:
      print("Insufficient Balance")
  elif choice == 3:
    if balance >= 50:
      balance = balance - 50
      print("Transaction successful")
    else:
      print("Insufficient Balance")
  elif choice == 4:
    if balance >= 100:
      balance = balance - 100
      print("Transaction successful")
    else:
      print("Insufficient Balance")
  elif choice == 5:
    if balance >= 500:
      balance = balance - 500
      print("Transaction successful")
    else:
      print("Insufficient Balance")
  elif choice == 6:
    amount = int(input("Please enter an amount to deposit:"))
    if amount > 0 and amount % 10 == 0:
      if balance >= amount:
        balance = balance - amount
        print("Transaction successful")
      else:
        print("Insufficient Balance")
    else:
      print("Invalid amount")
  else:
    print("Invalid option")
  while True:
    exit = input("Would you like to perform another  action? Y/N")
    if exit == "N":
      print("Exiting ATM")
      break
    elif exit == "Y":
      return menu()
    else:
      print("Invalid option, try again")

#Menu Function
def menu():
  while True:
      option = int(input("Please Select the operation you wish to use:\n"\
                         "1: Check Balance\n"
                         "2: Withdraw\n"
                         "3: Deposit\n"
                         "4: Quit\n"))
      if option == 1:
        return check_balance()
      elif option == 2:
        return withdraw()
      elif option == 3:
        return deposit()
      elif option == 4:
        print("Exiting ATM")
        break
      else:
        print("Invalid option, try agin")

#Access and starting program
#Pin code for ATM
pin = 1234
#Starting balance
balance = 10000
#Pin validation
Auth = False
while Auth == False:
  access = int(input("Please enter the pin to access the ATM:"))
  if access == pin:
    Auth = True
    menu()
  else:
    print("Invalid pin, try again")
```
